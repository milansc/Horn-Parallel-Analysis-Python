```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

def Hornpar ( num_simulations, var, row_size, x_df):
    
# num: Define number of desired random data simulations 
# var: Define number of original (manifest) variables that go into principal component analysis
# row_size: number of respondents/cases from your dataset
# x_df: DATAFRAME with SCALED manifest variables, THIS IS YOUR DATAFRAME FROM WHICH YOU CONDUCT PCA/FACTOR ANALYSIS
    
# LEAVE THIS LIST EMPTY, it will be used later    
    rn_eigen = []
    
#random numbers in normal distribution, this is what I chose, but parallel analysis is not that sensitive to distribution
# so you can choose another distribution type if you want
# row size = NO OF ROWS IN DATA FOR PRINCIPAL COMPONENTS ANALYSIS
# data for pca must be scaled, so we scale the random data with StandardScaler

    for i in range(num_simulations):
          rnormal = np.random.randn(row_size, var) 
          scaler = StandardScaler()
          scaler.fit(rnormal)
          rnormal_scaled = scaler.transform(rnormal)
          rnormal_pca = PCA().fit(rnormal_scaled)
          rn_eig_vals = rnormal_pca.explained_variance_
          rn_eigen.append(rn_eig_vals)
             
# now we make a list with all of the eigen_values from random data beggining with first simulation eigen values,
#then second simulation eigen values, until the last simulation
# LEAVE THESE LISTS EMPTY, they will be used later    
    rn_eigen_all = []
    rn_eigen_order = []
    for i in range(var):
          for j in range(num_simulations):
                rn_eigens_all = rn_eigen[j][i]
                rn_eigen_all.append(rn_eigens_all) 
  
      
            
# LEAVE THESE LISTS EMPTY, they will be used later       
    average_rn_eigens = []
    perc95_rn_eigens = []       
    
            
# now we arrange all of eigen values beggining with first eigen values of every simulation, than second eigen values, 
# until the last eigen values which are the last n values, where n is number of simulations
# we get average(median, 50 percentile) and 95 percentile of eigen value distribution for every order of eigen values
    
    for j in range(var):
            rn_eigen_order.append(rn_eigen_all[j*num_simulations:(j+1)*num_simulations])
            average_rn_eig = np.percentile(rn_eigen_order,50)
            perc95_rn_eig = np.percentile(rn_eigen_order,95)
            average_rn_eigens.append(average_rn_eig)
            perc95_rn_eigens.append(perc95_rn_eig)
            rn_eigen_order = []

# at the end we insert the actual eigen values from actual principal component analysis on real, manifest variables
#x_df is THE DATAFRAME with SCALED manifest variables, THIS IS YOUR DATAFRAME FROM WHICH YOU CONDUCT PCA/FACTOR ANALYSIS
    x_df_pca = PCA().fit(x_df)
    eig_vals = x_df_pca.explained_variance_
    eig_values = pd.DataFrame({'Actual eig_values' : eig_vals})
    rn_eigens = pd.DataFrame({'Avg. eigens of random numbers': average_rn_eigens, 
                          '95% eigens of random numbers' : perc95_rn_eigens})
    parallel_df = eig_values.join(rn_eigens)
    display(parallel_df)

# First output: 
#dataframe with actual eigenvalues from your data, average and 95 perc. of distribution of eigen values of random data

# you keep those components that have actual eigenvalues greater than 95% percentile of random data eigen_values

# Second output and RETURNED:
#showing components with eig values greater than 95% percentile of random components - this is the DESIRED NUMBER OF FACTORS
    
    parallel_criterium_df = parallel_df[parallel_df['Actual eig_values']>parallel_df['95% eigens of random numbers']]
    print("SELECTED COMPONENTS/FACTORS ACCORDING TO PARALLEL ANALYSIS IS : " , len(parallel_criterium_df))
    display(parallel_criterium_df)
    return(parallel_df, parallel_criterium_df)
    
# if you want to store the results into a variable, you must USE TWO VARIABLES, ONE FOR ALL EIGENVALUES, 
#SECOND ONLY FOR THOSE COMPONENTS THAT FIT THE PARALLEL ANALYSIS CRITERIUM
        
            
            
        


```
